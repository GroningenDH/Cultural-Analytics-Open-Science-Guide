### Solution to Q2

In this exercise, the formula will be used: the index of popularity of a film = rating x votes, which is then divided by the number of genres. To find out what sort of subgenres were popular throughout the Film Noir age, we made two additional variables. The first variable is the 'Score', which is created by multiplying 'Rating' with 'Votes'. The second variable is the 'Popularity Index', which is created by dividing the score by the number of distinct entries of genre. 

```python
film_noir_df['Votes'] = film_noir_df['Votes'].str.replace(r'[^\w\s]+', '')
```

```python
film_noir_df[['Votes']] = film_noir_df[['Votes']].astype(float)
```

```python
film_noir_df['index'] = film_noir_df[['Rating','Votes']].apply(lambda x:x['Rating']*x['Votes'],axis=1)
film_noir_df.head(5)
```

```python
# Score is made by adding all the Votes together by Genre. 
score_sum = film_noir_df.groupby('Genre')['index'].sum().reset_index(name = 'Score')
score_sum.head(10)
```

```python
sub_genre = film_noir_df.groupby(['Genre'])['Genre'].count().reset_index(name = 'Count').sort_values(['Count'], ascending = False)
sub_genre.head(10)
```

```python
df = score_sum.set_index('Genre').combine_first(sub_genre.set_index('Genre')).reset_index()
df['Popularity_Index'] = round((df['Score']/df['Count']),2)
df
```

```python
df = df.sort_values(by='Popularity_Index', ascending=False)
for_visual = df.head(20)
for_visual
```

### Solution to Q5b

```python
desc = film_noir_df['desc']
descriptions = []
for description in desc:
    if description.endswith("Â»"):
        description = description[:description.find("...                See full")]
    descriptions.append(description)

    
    
frequencies = {}
for description in descriptions:
    for word in description.split(" "):
        if word.lower() in frequencies:
            frequencies[word.lower()] += 1
        else:
            if word.lower() not in stop_w:
                frequencies[word.lower()] = 1
frequencies = sorted(frequencies.items(),key=lambda item:item[1],reverse=True)
frequencies
```

By completing the exercise now you are able to see the results from both methods are similar, and both ways of retrieving and displaying the frequencies are equally useful. Moreover, the results reveal the thematic patterns of the Film Noir genre.

```python
from glob import glob
from os.path import splitext
from os.path import basename
import os
import pandas as pd
```

```python
editions = []
for edition in glob('../texts_clean/*.txt'):
    editions.append(edition)
```

```python
editions_metadata = pd.read_csv('../editions.csv', encoding='utf-8', index_col=0)
editions_metadata
```

```python
newspapers = pd.read_csv('../newspapers.csv', encoding='utf-8', index_col=0)
newspapers
```

```python
editions_full = editions_metadata.merge(newspapers, on = "newspaper")
editions_full
```

```python
def create_subcorpus(feature):
    subcorpus = str(feature + "/corpus")
    if not os.path.exists(subcorpus):
        os.makedirs(subcorpus)
    for edition in editions:
        edition_id = int(os.path.splitext(basename(edition))[0])
        feature_lookup = str(editions_full.iloc[edition_id-1][feature])
        edition_id = str(edition_id)
        with open(edition, "r", encoding = "utf8") as oldfile, open(subcorpus + "/" + feature_lookup + "_" + edition_id + ".txt", "w", encoding = "utf8") as newfile:
            for line in oldfile:
                newfile.write(line)
        
for feature in ["newspaper", "ideology", "hq", "audience", "region", "format", "year", "foundedin"]:
    create_subcorpus(feature)
```
